```python

#  anova
from scipy import stats
fstat, pvalue = stats.f_oneway(df3['Year'], df3['Year'])
if pvalue < 0.05:
   print("Reject null hypothesis")

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import  r2_score
```

```python
categories = df3['Category'].unique() 
category_mapping = {c:i for i, c in enumerate(categories)}

df3['Category'] = df3['Category'].map(category_mapping)

publishers = df3['Publisher'].unique()
publisher_mapping = {p:i for i,p in enumerate(publishers)}

df3['Publisher'] = df3['Publisher'].map(publisher_mapping)

titles = df3['Title'].unique()
title_mapping = {c:i for i, c in enumerate(titles)}

df3['Title'] = df3['Title'].map(title_mapping)

authors = df3['Author'].unique()
author_mapping = {c:i for i, c in enumerate(authors)}

df3['Author'] = df3['Author'].map(author_mapping)
```

```python


 #inputs   
x = df3[['Price','Category','Publisher']].values
y = df3['Author']



# Split data
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.05, random_state= 2 )

# Create regression
reg = LinearRegression()

# Train model
reg.fit(X_train, y_train)   

# Make predictions
y_pred = reg.predict(X_test)

# R-squared

r2 = r2_score(y_test, y_pred)

print("R-squared:", r2)
print("Coefficients:", reg.coef_)
# Evaluate 
mse = mean_squared_error(y_test, y_pred) 
print("MSE:", mse)

```
